```python
import numpy as np # crucial for matplotlib
import matplotlib.pyplot as plt # used for plotting
```

```python
x = np.linspace(-100000, 100000, 1000000)
y = x**2 #easy way, calculates for all points (smaller number, highest number, number of points)
```

```python
plt.plot(x,y)
plt.xlabel("Argument $x$")
plt.xlabel("Value $y$")
```
